```python
# _*_ coding: UTF-8 _*_
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
import requests
```

```python
req = requests.get('https://www.billboard.com/charts/hot-100')
html = req.text
bsObject = BeautifulSoup(html, 'html.parser')
bsObject
```

```python
for n in range(0,100):
    span = bsObject.find_all("span", {"class":"chart-list-item__title-text"})
    title = span[n].get_text()
    div = bsObject.find_all("div", {"class":"chart-list-item__artist"})
    artist = div[n].get_text()
    title = title.replace("\n","")
    artist = artist.replace("\n","")
    print(str(n+1)+".",title," by "+artist)
```

```python
with open('Billboard_Ho1-100.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', 'artist'])
            
    for n in range(0,100):
        span = bsObject.find_all("span", {"class":"chart-list-item__title-text"})
        title = span[n].get_text()
        div = bsObject.find_all("div", {"class":"chart-list-item__artist"})
        artist = div[n].get_text()
        title = title.replace("\n","")
        artist = artist.replace("\n","")
        writer.writerow([n+1, title, artist])
```
